```python
# おまじない
from matplotlib import pyplot as plt

plt.rcParams['text.usetex'] = False
```

```python
import pandas as pd
df3 = pd.read_table('1e-3.dat',header = None)
r3 = df3[0]
phi3 = df3[1]

df4 = pd.read_table('1e-4.dat',header = None)
r4 = df4[0]
phi4 = df4[1]


df5 = pd.read_table('1e-5.dat',header = None)
r5 = df5[0]
phi5 = df5[1]

df6 = pd.read_table('1e-6.dat',header = None)
r6 = df6[0]
phi6 = df6[1]
```

```python
import numpy as np
a = 0.05 #外側電極の半径[m]。a < RMAXとすること
b = 2e-3 #内側電極の半径[m]。b < aとすること
t = 2e-3 #外側電極の厚さ[m]
Vmax = 10e3 #内側電極(高電圧電極)の電位[V]

r = np.linspace(b*100,a*100,100) #[cm]
phi_r = Vmax/10**3*np.log(a*100/r)/np.log(a/b) #[kV]
```

```python
from matplotlib import pyplot as plt

plt.plot(r3, phi3,'.', label = 'ERR_LIM = 1e-3')
plt.plot(r4, phi4,'.', label = 'ERR_LIM = 1e-4')
plt.plot(r5, phi5,'.', label = 'ERR_LIM = 1e-5',color = 'red',markersize = 8)
plt.plot(r6, phi6,'.', label = 'ERR_LIM = 1e-6',alpha = 0.3,markersize = 13)

plt.plot(r,phi_r,'+',label = 'theory',alpha = 0.5,color = 'blue')


plt.grid()

plt.legend()
plt.xlabel('r[cm]')
plt.ylabel('phi(r,1)[kV]')
plt.savefig('kadai1.png')
```

# 発展課題3_4(1)

```python
df = pd.read_table('kadai3_4_1.dat',header = None)
r = df[0]
z = df[1]
E = df[2]

```

```python
# PYTHON_MATPLOTLIB_3D_PLOT_01
from mpl_toolkits.mplot3d import Axes3D

# Figureを追加
fig = plt.figure(figsize = (8, 8))

# 3DAxesを追加
ax = fig.add_subplot(111, projection='3d')

# Axesのタイトルを設定
ax.set_title("E(r,z)", size = 20)

# 軸ラベルを設定
ax.set_xlabel("r[cm]", size = 14)
ax.set_ylabel("z[cm]", size = 14)
ax.set_zlabel("E[kV/cm]", size = 14)

# 軸目盛を設定
ax.plot(r,z,E,'.',alpha = 0.4,ms = 0.3)
plt.savefig('kadai3_4_1.png')
```

```python

import matplotlib.cm as cm

fig = plt.figure()
ax = fig.add_subplot(111)

H = ax.hist(E, bins=40, cmap=cm.jet)
ax.set_title('1st graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.colorbar(H[3],ax=ax)
plt.show()
```
